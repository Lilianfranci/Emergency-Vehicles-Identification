```python
from keras.preprocessing.image import ImageDataGenerator # Keras image data generator is used for the generation of images in batch
```

```python
image_gen = ImageDataGenerator(
    rotation_range=20,# specifies the rotation to be 20%
    width_shift_range=0.1, # specifies the width shift of 10%
    height_shift_range=0.1,# specifies the height shift of 10%
    shear_range=0.20, # crops part of the image
    zoom_range=0.20, # #zooms the image by 20%
    fill_mode='nearest', # fills using the nearest pixel
    horizontal_flip=True, # Secifies the horizontal flip
    rescale=1/255  # scales the image
)
```

```python
train_generator=image_gen.flow_from_dataframe(dataframe=train_df[:1150], # specify the dataset used for training
                                            directory=train_path, # specify the path to the directory
                                            x_col='image_names', # specify the names of the images
                                            y_col='emergency_or_not', # specifies the class labels
                                            class_mode='binary', # specifies the kind of classification
                                            target_size=(224,224), # specifies the dimension to resize the img
                                            batch_size=50, #defines the batch size
                                            validate_filenames=True
                                            )
test_generator=image_gen.flow_from_dataframe(dataframe=test_df[:1150], # specify the dataset used for training
                                            directory=test_path, # specify the path to the directory
                                            x_col='image_names', # specify the names of the images
                                            y_col='emergency_or_not', # specifies the class labels
                                            class_mode='binary', # specifies the kind of classification
                                            target_size=(224,224), # specifies the dimension to resize the img
                                            batch_size=50, #defines the batch size
                                            validate_filenames=True
                                            )
validation_generator=image_gen.flow_from_dataframe(dataframe=train_df[1150:], # specify the dataset used for training
                                            directory=train_path, # specify the path to the directory
                                            x_col='image_names', # specify the names of the images
                                            y_col='emergency_or_not', # specifies the class labels
                                            class_mode='binary', # specifies the kind of classification
                                            target_size=(224,224), # specifies the dimension to resize the im
                                            batch_size=50, #defines the batch size
                                            validate_filenames=True
                                            )
```

### Constructor stage

```python
model = Sequential()
model.add(Conv2D(filters=32,kernel_size=(3,3),input_shape=(224,224,3),activation = 'relu',padding = 'same'))
#model.add(MaxPooling2D(pool_size=(2,2)))
#model.add(Conv2D(filters=32,kernel_size=(3,3),input_shape=(224,224,3),activation = 'relu', padding = 'valid',s
#model.add(MaxPooling2D(pool_size=(2,2)))
#model.add(Conv2D(filters=32,kernel_size=(3,3),input_shape=(224,224,3),activation = 'relu'))
#model.add(MaxPooling2D(pool_size=(2,2)))
#model.add(BatchNormalization())
model.add(Flatten())
model.add(Dense(50,activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(1,activation = 'sigmoid'))
```

### Compilation stage

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = 'accuracy')
```

```python
model.summary()
```

### Training stage

```python
history=model.fit_generator(train_generator,epochs = 5,validation_data = validation_generator)
```

```python
## checking for overfitting
history_df = pd.DataFrame(history.history)
plt.figure(dpi=200, figsize = (10,3))
plt.plot(history_df['loss'],
label='training_loss')
plt.plot(history_df['val_loss'],
label='val_loss')
plt.xlabel('Loss')
plt.ylabel('Epochs')
plt.legend()
plt.show()
```

### Evaluation stage

```python
prediction = model.predict(x=test_generator, verbose=0)
prediction = prediction>0.5
```

```python
#confusion matrix

conMax = confusion_matrix(y_true = test_generator.classes, y_pred = prediction)
conMax
```
